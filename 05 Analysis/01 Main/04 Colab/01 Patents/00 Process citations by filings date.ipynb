# Load and combine


```python
#@title Load and merge
import pandas as pd

dtype_spec = {0: str, 3: str, 6: str}

citations = pd.read_csv("/content/drive/MyDrive/PhD Data/00 Raw data from PV/g_us_patent_citation.tsv/g_us_patent_citation.tsv", sep = "\t", dtype=dtype_spec)
filings = pd.read_csv("/content/drive/MyDrive/PhD Data/00 Raw data from PV/g_application.tsv/g_application.tsv", sep = "\t", low_memory=False)

citations.set_index('patent_id', inplace=True)
filings.set_index('patent_id', inplace=True)

## Keep relevant columns
cols = ['citation_patent_id', 'citation_category']
citations = citations[cols]

cols = ['filing_date']
filings = filings[cols]
```

```python
print(citations.columns)
print(filings.columns)
```

```python

# Merge on patent_id
merged_df = citations.join(filings, how="inner")

merged_df = merged_df.reset_index()
merged_df.head()
```

```python
## Filter columns & rename and save


# Save as pickle
merged_df.to_pickle('/content/drive/MyDrive/PhD Data/08 Citations/03 Patent citations - raw, filing.pickle')
```
