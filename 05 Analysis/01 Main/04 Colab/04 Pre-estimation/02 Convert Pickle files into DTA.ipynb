# Save files into Colab

```python
import api.process_matches as processing
import importlib
importlib.reload(processing)
import pandas as pd
import dill as pickle
from tqdm import tqdm
import glob

# Take all the PKL files in the directory and save them as DTA
# Input directory containing the old matched pickles
INPUT_DIR = "/content/drive/MyDrive/PhD Data/12 Sample Final/actual results/citation_no_exact_match_on_grantyear/"

# Output directory for final samples
OUTPUT_DIR = INPUT_DIR

# Glob pattern to find matched samples
INPUT_PATTERN = "Sample - *10matches.pkl"

# Columns to keep - treated_id control_id t0 rel_q quarter citations_treated citations_control lambda_val ult_parent deal_id
COLUMNS = ["treated_id", "control_id", "rel_q", "quarter", "citations_treated", "citations_control", "lambda_val", "ult_parent", "deal_id", "acq_date", "resold_date", "mahalanobis_distance_scaled",  "cosine_distance_scaled",  "hybrid_distance"]

for filepath in tqdm(glob.glob(os.path.join(INPUT_DIR, INPUT_PATTERN)), desc="Processing samples"):
        # Retrieve filename
        filename = os.path.basename(filepath)

        # Load the sample
        with open(filepath, 'rb') as f:
            sample = pickle.load(f)

        # Drop cases where cosine distance is exactly 0 - happens for continuation in part patents
        aux = sample[sample['cosine_distance'] == 0]
        if len(aux) == 0:
            print(f"No cases to drop for {filename}, skipping this file as it already exists in drive.")
            continue
        else:
            print(f"Dropping {len(aux)} cases out of {len(sample)} where cosine distance is exactly 0 for {filename}")
        sample = sample[sample['cosine_distance'] != 0]

        # Filter relevant columns
        sample = sample[COLUMNS]

        # Make quarter column string
        sample["quarter"] = sample["quarter"].astype(str)

        # Save as DTA
        sample.to_stata(os.path.join(OUTPUT_DIR, filename.replace(".pkl", ".dta")), write_index=False)
        #print(f"Saved {filename.replace('.pkl', '.dta')}, with shape {sample.shape}")
```
