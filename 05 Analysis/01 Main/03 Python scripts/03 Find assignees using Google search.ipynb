```python
## TFIDF matches
from api.google_search_checker import perform_google_search

local_counter = 0
totalcount = len(match_tfidf)
match_rest['google_matched'] = None
## Run over the target assignees dict to run the Google Search
for index, row in match_rest.iterrows():
    
    ## Get the name  of the target 
    target = row['small_dataset_name']
    match = [row['match'],]
    
    ## Perform the google search
    google_matches, apicallcount = perform_google_search(target, match)

    if match == google_matches:
        print(f"{target} has a match in TF-IDF")
        #row['google_matched'] = google_matches
        match_rest.at[index, 'google_matched'] = 1

    ## Print progress
    local_counter += 1
    print(f"Google search progress : {local_counter/totalcount}")

    ## Print API counts
    if apicallcount % 50 == 0:
        print("API call count is: ", apicallcount)

```
