```python
import xml.etree.ElementTree as ET
import requests
import pandas as pd
import datetime

# List of transactions as (Assignor, Assignee) tuples
transactions = [
    ("INTERNATIONAL BUSINESS MACHINES CORPORATION", "FACEBOOK INC."),

    ("INTERNATIONAL BUSINESS MACHINES CORPORATION", "GOOGLE INC."),
    ("INTERNATIONAL BUSINESS MACHINES CORPORATION", "MICROSOFT CORPORATION"),
    ("INTERNATIONAL BUSINESS MACHINES CORPORATION", "NUANCE COMMUNICATIONS, INC."),

    ("ROCKSTAR BIDCO, LP", "APPLE INC."),
    ("ROCKSTAR BIDCO, LP", "MICROSOFT CORPORATION"),

    ("EASTMAN KODAK COMPANY", "APPLE INC."),

    ("INTEL CORPORATION", "APPLE INC."),



    ("MICROSOFT CORPORATION", "FACEBOOK INC."),

    ("AOL INC.", "FACEBOOK, INC."),
    ("AOL, INC.", "FACEBOOK, INC."),
    ("AOL LLC", "FACEBOOK, INC."),
    ("AMERICA ONLINE INC.", "FACEBOOK, INC."),

    ("AOL INC.", "MICROSOFT CORPORATION"),
    ("AOL LLC", "MICROSOFT CORPORATION"),
    ("AOL, INC.", "MICROSOFT CORPORATION"),
    ("AMERICA ONLINE INC.", "MICROSOFT CORPORATION"),
    ("AOL LLC.", "MICROSOFT CORPORATION"),
    ("AOL ADVERTISING INC.", "MICROSOFT CORPORATION"),

    ("AOL INC.", "MICROSOFT TECHNOLOGY LICENSING LLC"),
    

    ("MICROSOFT TECHNOLOGY LICENSING LLC", "FACEBOOK INC."),

    ("MOTOROLA MOBILITY LLC", "GOOGLE TECHNOLOGY HOLDINGS LLC"),
    ("MOTOROLOA MOBILITY LLC", "GOOGLE TECHNOLOGY HOLDINGS LLC"),
     
]

# Initialize a list to store all extracted data
all_patent_records = []

# Loop over each assignor-assignee pair
for assignor, assignee in transactions:
    # Construct the API URL for each transaction
    url = f"https://assignment-api.uspto.gov/patent/advancedSearch?fields=id&rows=10000&ownerName={assignee.replace(' ', '%20')}&priorOwnerName={assignor.replace(' ', '%20')}"
    print(url)
    # Fetch XML data
    response = requests.get(url)
    xml_data = response.content

    # Parse XML
    root = ET.fromstring(xml_data)

    # Locate all <doc> elements
    docs = root.findall(".//doc")

    # Initialize a list for this transaction
    transaction_records = []

    # Loop over each document
    for doc in docs:
         # Extract recorded date
        execution_date_elem = doc.find("arr[@name='patAssignorExDate']")
        execution_date = execution_date_elem.findall("date")[0].text

        # Extract patent numbers from <arr name="patNum">
        patent_numbers = []
        for arr in doc.findall("arr[@name='patNum']"):
            for str_elem in arr.findall("str"):
                if str_elem.text:
                    patent_numbers.append(str_elem.text)

        # Extract patent numbers from <lst name="patNum">
        for lst in doc.findall("lst[@name='patNum']"):
            for int_elem in lst.findall("int"):
                if int_elem.text:
                    patent_numbers.append(int_elem.text)

        # Store the result if there are patent numbers
        if execution_date and patent_numbers:
            for patent in patent_numbers:
                transaction_records.append({
                    "Assignor": assignor,
                    "Assignee": assignee,
                    "Execution Date": execution_date,
                    "Patent Number": patent
                })

    # Append to global list
    all_patent_records.extend(transaction_records)

    # Print the number of patents found for this transaction
    print(f"Transaction: {assignor} → {assignee}, Patents Found: {len(transaction_records)}")
# Convert to DataFrame
df = pd.DataFrame(all_patent_records)

# Define the folder path where you want to save the CSV file
folder_path = r"C:\Users\amusali\Desktop\uc3m PhD\05 Analysis\01 Main\03 Python scripts\02 Find off deal patents\out"  # Change this to your desired folder path

# Get the current timestamp in YYYYMMDD_HHMMSS format
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")

# Create a filename with the timestamp
csv_filename = f"{folder_path}\\all_patents_transferred_{timestamp}.csv"

# Save DataFrame as CSV
df.to_csv(csv_filename, index=False)

print(f"Extracted {len(all_patent_records)} total patent records and saved them to '{csv_filename}'")
```
