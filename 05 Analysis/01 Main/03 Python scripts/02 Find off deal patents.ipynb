```python
import xml.etree.ElementTree as ET
import requests

# URL from the API
url = "https://assignment-api.uspto.gov/patent/advancedSearch?fields=id&ownerName=FACEBOOK%2C%20INC.&priorOwnerName=INTERNATIONAL%20BUSINESS%20MACHINES%20CORPORATION"

# Get the XML response from the API
response = requests.get(url)
xml_data = response.content

# Parse the XML data
root = ET.fromstring(xml_data)

# Locate all <arr> elements with the name attribute "patNum"
patnum_arrs = root.findall(".//arr[@name='patNum']")

# Initialize a list to hold the patent numbers
patent_numbers = []

# Loop through each <arr> element and extract <str> elements
for arr in patnum_arrs:
    for str_elem in arr.findall("str"):
        if str_elem.text:
            patent_numbers.append(str_elem.text)

print("Extracted Patent Numbers:")
for patent in patent_numbers:
    print(patent)

print(f"Length of patent numbers {len(patent_numbers)}")

```
