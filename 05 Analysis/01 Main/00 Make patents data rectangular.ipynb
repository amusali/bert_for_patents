## Load patents that has ben used as a sample 

```python
import dill as pickle
from datetime import datetime

# Set the path to the pickle file
path = r"C:\Users\amusali\Desktop\uc3m PhD\05 Analysis\01 Main\00 Python data\03 Cleaned patents\all_patents.pkl"

# Load the pickle file
with open(path, 'rb') as file:
    patents = pickle.load(file)

# Display the loaded patents
print(len(patents))

## Motorola patents
path_motorola = r"C:\Users\amusali\Downloads\Motorola Mobility LLC_2012-05-21_before_0135.pkl"
with open(path_motorola, "rb") as file:
    motorola_patents = pickle.load(file)

for patent in motorola_patents:
    patent.date_acquired = datetime(2012, 5, 21)

# Display the loaded patents
print(len(motorola_patents))

## All patents
all_patents = patents + motorola_patents
```

### Get the patents into a rectangular form

```python
import pandas as pd
import numpy as np

data = []
    
# Convert the patents dictionary to a pandas DataFrame
for patent in all_patents:
    # Check if the patent has a closest patent
    if patent.closest_patent is None:
        continue
    
    # Extract the relevant information
    acquisition_date = patent.date_acquired
    granted = False
    matched_pair_id = patent.patent_id

    # Get the row for each acquired patent x closest patent
    row = {
        # Data for acquired patent (treatment)
        'patent_id': patent.patent_id,
        'application_date': patent.date_application,
        'grant_date': patent.date_granted,
        'Assignees': patent.assignee_organization,
        'assignee_country': patent.assignee_country,
        'forward_citations': patent.forward_citations,
        'cpc_type_current': patent.tech_field_group,
        'cpc_subclass_current': patent.tech_field_group_id,
        'cpc_group_current': patent.tech_field_subgroup_id,
        

        # Data for closest patent (control)
        'patent_id_closest': patent.closest_patent.patent_id,
        'application_date_closest': patent.closest_patent.date_application,
        'grant_date_closest': patent.closest_patent.date_granted,
        'Assignees_closest': patent.closest_patent.assignee_organization,
        'assignee_country_closest': patent.closest_patent.assignee_country,
        'forward_citations_closest': patent.closest_patent.forward_citations,
        'cpc_type_current_closest': patent.closest_patent.tech_field_group,
        'cpc_subclass_current_closest': patent.closest_patent.tech_field_group_id,
        'cpc_group_current_closest': patent.closest_patent.tech_field_subgroup_id,

        # Data for both types
        'DateEffective': patent.date_acquired,
        'cosine_similarity_with_closest_patent': patent.cosine_similarity_with_closest_patent,
        'euclidean_distance_to_closest_patent': patent.euclidean_distance_to_closest_patent,
    }
    data.append(row)

# Generate a dataframe from the data
df = pd.DataFrame(data)
   
```
